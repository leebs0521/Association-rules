```python
processing_dataset=dataset.select_dtypes('object').drop(columns=['CODE_GENDER'])
```

```python
processing_dataset
```

```python
processing_dataset.columns
```

```python
new_columns = ['Target',
    'Contract type',
    'Flag own car',
    'Flag own house',
    'Children',
    'Income',
    'Credit loan',
    'Income type',
    'Education type',
    'Family status',
    'Housing type',
    'Flag work phone',
    'Weekday appr process start',
    'Organization type'
]

processing_dataset.columns = new_columns
```

```python
processing_dataset
```

```python
reject_data = processing_dataset[processing_dataset.Target=='Reject']
```

```python
accept_data = processing_dataset[processing_dataset.Target=='Accept'].sample(n=24825)
```

```python
accept_data
```

```python
reject_data
```

```python
mydataset = pd.concat([reject_data, accept_data]).sort_index()
```

```python
mydataset.Target.value_counts()
```

```python
mydataset.head()
```

```python
mydataset.to_csv('./mydataset.csv', mode='w', index=False)
```

```python
transaction_data = mydataset.Target
for col in mydataset.columns:
    if col != 'TARGET':
        transaction_data = transaction_data + ',' + mydataset[col]
```

```python
transaction_data
```

```python
transaction_df = pd.DataFrame(transaction_data, columns=['data'])
```

```python
transaction_df
```
